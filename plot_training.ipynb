```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

import seaborn as sns
```

```python
barcolor = ['orange', 'blue' ]
```

# T50

## Reward

```python
#==========================================Read Data
dqn_lin = pd.read_csv('training_plot/dqn_rew_t50.csv')
#data['ilp_qoe'] = data['ilp_qoe'].abs() #few results are actually positive but showing negetive
# dqn_exp = pd.read_csv('training_plot/dqn_exp_rew_t50.csv')
```

```python
fig1, ax1 = plt.subplots(linewidth=18) #figsize=(10,7) add to increase fig size
legend = [r'Reward, $\Gamma=50$', r'Exponential Reward, $\Gamma=50$']

plt.locator_params(axis='x', nbins=5)


sns.lineplot(x=dqn_lin.Step, y=dqn_lin.Value, ax = ax1, linewidth=0.5, linestyle='--', color=barcolor[0])
# sns.lineplot(x=dqn_exp.Step, y=dqn_exp.Value, ax = ax1, linewidth=0.5, linestyle='--', color=barcolor[1])


ax1.set_ylabel('Average Reward', fontsize=12, fontweight="bold")

ax1.set_xlabel("Number of Training Steps", fontsize=12, fontweight="bold")

ax1.set_ylim([-5, 15])

ax1.tick_params(axis='both', labelsize=12)
ax1.legend(legend, loc='upper left', bbox_to_anchor=(0, 1), prop = {'weight':'bold', 'size':  10})
ax1.grid(True)

# plt.xticks(rotation = 0)

fig1.set_figwidth(4)
fig1.set_figheight(3)

fig1.savefig("training_plot/dqn_reward_t50.pdf",dpi=100, bbox_inches = "tight")
```
